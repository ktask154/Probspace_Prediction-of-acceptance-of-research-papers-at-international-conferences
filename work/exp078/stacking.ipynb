```python
train = pd.read_csv(os.path.join(INPUT_DIR,"train_data.csv"))
test = pd.read_csv(os.path.join(INPUT_DIR,"test_data.csv"))
sub = pd.read_csv(os.path.join(INPUT_DIR,"submission.csv"))
```

```python
for cfg in [CFG63, CFG64, CFG65, CFG66, CFG68, CFG69, CFG70, CFG71, CFG72, CFG74, CFG75, CFG76, CFG77]:
#for cfg in [CFG64, CFG65, CFG71, CFG76]:
  oof_df = pd.read_pickle(cfg.path+'oof_df.pkl')
  oof = pd.DataFrame()
  for fold in cfg.trn_fold:
    _df = oof_df.loc[oof_df.kfold==fold]
    thresh = cfg.fold_thresh[fold]
    pred = _df["pred"]
    _df[f"{cfg.model.replace('/', '-')}_pred_label"] = (pred>thresh).astype(int)
    oof = pd.concat([oof,_df])
  train = train.merge(oof[["id", f"{cfg.model.replace('/', '-')}_pred_label"]], how="left", on="id")
  
  pred_df = pd.read_csv(os.path.join(OUTPUT_DIR, f"submition_seed{cfg.seed}_ver{cfg.ver}_{cfg.model.replace('/', '-')}.csv"))
  pred_df.columns = ["id", f"{cfg.model.replace('/', '-')}_pred_label"]
  test = test.merge(pred_df[["id", f"{cfg.model.replace('/', '-')}_pred_label"]], how="left", on="id")
```

```python
oof_features = train.loc[:,"microsoft-deberta-v3-base_pred_label":].columns.to_list()
```

```python
train["num_title_word"] = train["title"].apply(lambda x: len(x.split()))
train["num_abst_word"] = train["abstract"].apply(lambda x: len(x.split()))

test["num_title_word"] = test["title"].apply(lambda x: len(x.split()))
test["num_abst_word"] = test["abstract"].apply(lambda x: len(x.split()))

train["keywords"] = train["keywords"].fillna("")
test["keywords"] = test["keywords"].fillna("")

train["num_keywords"] = train["keywords"].apply(lambda x : x.count(","))
test["num_keywords"] = test["keywords"].apply(lambda x : x.count(","))

train["num_keywords"] = train["num_keywords"].apply(lambda x : x if x==0 else x+1)
test["num_keywords"] = test["num_keywords"].apply(lambda x : x if x==0 else x+1)

def count_key_words(text, key_words):
    count = 0
    for key_word in key_words:
        if key_word in text:
            count += 1
    return count

def have_key_woreds(df):
  has_title_keywords, has_abst_keyworeds = [],[]
  for index, data in df.iterrows():
    title_text = data["title"]
    abst_text = data["abstract"]
    key_words = data["keywords"].split(",")
    num_title_keywords = count_key_words(title_text, key_words)
    num_abst_keywords = count_key_words(abst_text, key_words)
    has_title_keywords.append(num_title_keywords)
    has_abst_keyworeds.append(num_abst_keywords)
  return has_title_keywords,has_abst_keyworeds

train_has_title_keywords, train_has_abst_keywords = have_key_woreds(train)
test_has_title_keywords, test_has_abst_keywords = have_key_woreds(test)

train["have_title_keywords"] = train_has_title_keywords
train["have_abst_keywords"] = train_has_abst_keywords
test["have_title_keywords"] = test_has_title_keywords
test["have_abst_keywords"] = test_has_abst_keywords
```

```python
features = oof_features + ["num_title_word", "num_abst_word", "num_keywords", "have_title_keywords", "have_abst_keywords"]
```

```python
oof_df = pd.read_pickle(CFG63.path+'oof_df.pkl')
train = train.merge(oof_df[["id","kfold"]], how="left", on="id")
```

```python
ctb_oof = pd.DataFrame()
clf = CatBoostClassifier(custom_loss=['Accuracy'],
                           random_seed=42)
test_preds = np.zeros((len(test), 10))
for fold in range(10):
  print("="*15,f" CV FOLD {fold} ","="*15)
  train_fold = train.loc[train.kfold!=fold].reset_index()
  valid_fold = train.loc[train.kfold==fold].reset_index()

  train_X = train_fold[features]
  valid_X = valid_fold[features]
  test_X = test[features]
  train_y = train_fold["y"]
  valid_y = valid_fold["y"]

  clf.fit(train_X, train_y,
           eval_set=[(valid_X, valid_y)],
           early_stopping_rounds=50,
          cat_features = oof_features,
           verbose = 200)
  
  pred = clf.predict_proba(valid_X)[:,1]
  test_pred = clf.predict_proba(test_X)[:,1]

  valid_fold["pred"] = pred

  ctb_oof = pd.concat([ctb_oof, valid_fold])
  test_preds[:, fold] = test_pred
```

```python
def get_score(labels, outputs):
    y_pred = outputs
    y_true = labels
    best_score = 0
    best_thresh = 0.5
    for thresh in np.arange(0.1, 0.80, 0.01):
        thresh = np.round(thresh, 2)
        score = accuracy_score(y_true, (y_pred>thresh).astype(int))
        #print("Accuracy score at threshold {0} is {1}".format(thresh, score))
        if score > best_score:
          best_score = score
          best_thresh = thresh
    return accuracy_score(y_true, (y_pred>best_thresh).astype(int)), best_thresh
```

```python
score, thresh = get_score(ctb_oof.y, ctb_oof.pred)
```

```python
ctb_pred = np.mean(test_preds, axis=1)
test["ctb_pred"] = ctb_pred
test["pred"] = test["ctb_pred"].apply(lambda x : 1 if x>thresh else 0)
```
