```python
import os

DIR = "/content/drive/MyDrive/Competitions/probspace/研究論文の国際学会採択予測"
INPUT_DIR = os.path.join(DIR,"input")
OUTPUT_DIR = os.path.join(DIR,"output")

if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)
```

```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

from transformers import (AutoModel,AutoModelForMaskedLM, 
                          AutoTokenizer, LineByLineTextDataset,
                          DataCollatorForLanguageModeling,
                          Trainer, TrainingArguments)
```

```python
train = pd.read_csv(os.path.join(INPUT_DIR,"train_data.csv"))
test = pd.read_csv(os.path.join(INPUT_DIR,"test_data.csv"))

data = pd.concat([train,test])
data["texts"] = data["title"] + "[SEP]" + data["abstract"]  
data['texts'] = data['texts'].apply(lambda x: x.replace('\n',''))

text  = '\n'.join(data.texts.tolist())

with open(os.path.join(OUTPUT_DIR,'text.txt'),'w') as f:
    f.write(text)
```

```python
model_name = 'google/electra-base-discriminator'
model = AutoModelForMaskedLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.save_pretrained(os.path.join(OUTPUT_DIR,'clrp_google_electra_base_discriminator_epoch10'));
```

```python
train_dataset = LineByLineTextDataset(
    tokenizer=tokenizer,
    file_path=os.path.join(OUTPUT_DIR,'text.txt'), #mention train text file here
    block_size=256)

valid_dataset = LineByLineTextDataset(
    tokenizer=tokenizer,
    file_path=os.path.join(OUTPUT_DIR,'text.txt'), #mention valid text file here
    block_size=256)

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=True, mlm_probability=0.15)

training_args = TrainingArguments(
    output_dir=os.path.join(OUTPUT_DIR,"clrp_google_electra_base_discriminator_epoch10_chk"), #select model path for checkpoint
    overwrite_output_dir=True,
    num_train_epochs=10,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    evaluation_strategy= 'steps',
    save_total_limit=2,
    eval_steps=500,
    metric_for_best_model='eval_loss',
    greater_is_better=False,
    load_best_model_at_end =True,
    prediction_loss_only=True,
    report_to = "none")

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset)
```

```python
trainer.train()
trainer.save_model(os.path.join(OUTPUT_DIR,'clrp_google_electra_base_discriminator_epoch10'))
```
